```python
#internshala
import requests
from bs4 import BeautifulSoup

# Set the URL of the product listing page
base_url = "https://www.amazon.in/s"
params = {
    "k": "bags",
    "crid": "2M096C61O4MLT",
    "qid": "1653308124",
    "sprefix": "ba%2Caps%2C283",
    "ref": "sr_pg_1"
}

# Scrape 20 pages
for page in range(1, 21):
    print(f"Scraping page {page}...")

    params["ref"] = f"sr_pg_{page}"
    response = requests.get(base_url, params=params)
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the product containers
    products = soup.find_all("div", class_="sg-col-inner")

    # Extract the required information from each product
    for product in products:
        # Product URL
        url = product.find("a", class_="a-link-normal")
        if url:
            product_url = "https://www.amazon.in" + url["href"]
        else:
            product_url = "N/A"

        # Product Name
        name = product.find("span", class_="a-size-medium")
        product_name = name.text.strip() if name else "N/A"

        # Product Price
        price = product.find("span", class_="a-offscreen")
        product_price = price.text.strip() if price else "N/A"

        # Rating
        rating = product.find("span", class_="a-icon-alt")
        product_rating = rating.text.strip() if rating else "N/A"

        # Number of reviews
        reviews = product.find("span", class_="a-size-base")
        product_reviews = reviews.text.strip() if reviews else "N/A"

        # Print or store the scraped data as per your requirement
        print("Product URL:", product_url)
        print("Product Name:", product_name)
        print("Product Price:", product_price)
        print("Rating:", product_rating)
        print("Number of Reviews:", product_reviews)
        print("---------------------------")
```
